```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

# Plotting a Vector

```python
plt.quiver(0,0,4,5)
```

```python
plt.quiver(0,0,4,5,scale_units="xy",angles="xy",scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy",angles="xy",scale=1 , color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy",angles="xy",scale=1,color='b')
plt.quiver(0,0,-3,-6,scale_units="xy",angles="xy",scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Addition of two vectors

```python
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])

sum = vec1 + vec2 
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units="xy",angles="xy",scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units="xy",angles="xy",scale=1,color='r')
plt.quiver(0,0,5,1,scale_units="xy",angles="xy",scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

# Subtraction for two vectors

```python
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])

sub = vec1 - vec2 
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units="xy",angles="xy",scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units="xy",angles="xy",scale=1,color='r')
plt.quiver(0,0,-1,5,scale_units="xy",angles="xy",scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```
